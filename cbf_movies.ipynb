## Content-Based Filtering

```python
import pandas
movies = pandas.read_csv("movies.csv", sep=",")
```

```python
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words="english")
movies["overview"] = movies["overview"].fillna("")
```

```python
tfidf_matrix = tfidf.fit_transform(movies["overview"])
```

```python
tfidf_matrix
```

```python
pandas.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

```python
tfidf_matrix.shape
```

```python
movies["overview"]
```

## Similarity matrix

```python
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
```

## Find the most similar movies to a certain movie

```python
movie_title = "John Carter"
```

```python
idx = movies.loc[movies["title"]==movie_title].index[0]
idx
```

#### The key parameter tells sorted() to use the second element [1] of each (index, score) tuple
#### as the sorting criterion. The lambda function extracts the score value for comparison.
#### reverse=True orders from highest to lowest similarity

```python
scores = list(enumerate(similarity_matrix[idx]))
scores
```

```python
sorted(scores, key=lambda x: x[1], reverse=True)
```

#### Convert the NumPy array of similarity scores to a native Python list
#### The tolist() method transforms each np.float64 element into a standard float,
#### allowing sorted() to return clean numeric values without the NumPy type wrapper

```python
sorted(similarity_matrix[5].tolist(), reverse=True)
```

```python
scores
```

```python
movies_indices = [tpl[0] for tpl in scores[1:4]]
movies_indices
```

```python
list(movies["title"].iloc[movies_indices])
```

```python
def similar_movies(movies_title, nr_movies):
    idx = movies.loc[movies["title"]==movies_title].index[0]
    scores = list(enumerate(similarity_matrix[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    movies_indices = [tpl[0] for tpl in scores[1:nr_movies+1]]
    similar_titles = list(movies["title"].iloc[movies_indices])
    return similar_titles
```

```python
similar_movies("Kung Fu Panda 3", 3)
```
